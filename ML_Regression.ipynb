```python
# Hypothesis Testing
```

```python
# Null Hypothesis(Ho)= Sales value will not be affected by number of custmers
# Alternate Hypothesis(Ha)=Sales value will be affected by number of customers
```

```python
X1=df.iloc[:,3]
print(X1)
```

```python
X2=df1.iloc[:,3]
print(X2)
X2.isnull().count()
```

```python
import statistics
```

```python
X1mean=statistics.mean(X1)
print(X1mean)
```

```python
std_dev=statistics.stdev(X1)
print(std_dev)
```

```python
std_dev1=1155.25
```

```python
import numpy as np
```

```python
from numpy import sqrt,abs,round
from scipy.stats import norm
over_sigma=sqrt(std_dev**2/len(X1)+std_dev**2/len(X2))
Z_value=(X1mean-X2mean)/over_sigma
pval=2*(1-norm.cdf(abs(Z_value)))
print(pval)
```

```python
# pval is less than alpha(0.05) so reject the null hypothesis
# Number of customers affects the sales
```
